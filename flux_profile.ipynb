```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as mcolors
import matplotlib as mpl

from eyring_model import *
```

```python
# some parameters
n_paths = 2000
n_jumps = 200

x_px = 40
y_px = int(n_paths/40)

T = 300
dH_barrier = 3.5
dS_barrier = -9/T
dH_sigma = dH_barrier/3
dS_sigma = -dS_barrier/3
```

```python
dist = 'normal'
params = {'mu'  : np.array([dH_barrier, dS_barrier]),
         'cov' : np.array([[dH_sigma**2,0],
                           [0,dS_sigma**2]])}

# generate the barriers
model = EyringModel(T=T)
for n in range(n_paths):
    model.add_Path(n_jumps=n_jumps, area=model.area/n_paths)
    model.paths[n].generate_membrane_barriers(dist=dist, dist_params=params, multi=True)


# save the permeabilities as a cross-section
im = np.zeros((x_px, y_px))
n = 0    
for i in range(x_px):
    for j in range(y_px):
        im[i,j] = model.paths[n].calculate_permeability()
        n += 1


# plot the image
fig, ax = plt.subplots(1,1, figsize=(8,6))

model.calculate_permeability()

cmap = 'viridis'
norm = mcolors.Normalize(model.permeabilities.min(),model.permeabilities.max())

ax.imshow(im)
cb = fig.colorbar(plt.cm.ScalarMappable(norm=norm, cmap=cmap), ax=ax)
cb.set_label('Permeability ($L/m^2/h$)')
ax.set_xlabel('X (paths)')
ax.set_ylabel('Y (paths)')
ax.set_title(f'Normally distributed barriers, $\sigma_P = {model.permeabilities.std():.4e}$');
```
